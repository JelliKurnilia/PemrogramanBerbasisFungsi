Membuat program Simple Password encryption, dengan ketentuan
1. Input password bertipe string dengan batasan panjang password
100 karakter
2. Setiap karakter dari string tersebut diubah menjadi ASCII value
3. Kemudian setiap ASCII value tersebut dikembangkan menjadi 3
value dengan dilakukan operasi berikut:

```
1. Value pertama dibagi dengan integer 26 kemudian
dibulatkan dan dijumlahkan dengan 80 kemudian diubah
kembali menjadi karakter
2. Value kedua dihitung sisa bagi dengan 26 kemudian
dijumlahkan dengan 80 kemudian diubah kembali menjadi
karakter
3. Value ketiga adalah karakter + jika value pertama lebih
besar dari value kedua, atau value ketiga adalah karakter -
jika value pertama tidak lebih besar dari value kedua
```
4. Deretkan value pertama, kedua dan ketiga untuk setiap karakter
dalam password tersebut sehingga didapatkan password dalam
bentuk baru


## **3a**. Bantulah user tersebut dengan membuatkan sebuah program yang secara otomatis mengubah password yang di input menjadi terenkripsi sesuai aturan tersebut!

````python
def enkripsi(password):                                  # membuat sintax fungsi 'enkripsi'
  if len(password) > 100 or len(password) == 0:          # 1. Pengecekan panjang  password dengan batasan 100 karakter
    print("Enter 1-100 characters")
  else:
    ascii = [str(ord(i)) for i in password]              # 2. Mengubah setiap karakter dari string tersebut menjadi ASCII value
    enkri = ''.join(chr((int(j)//26)+80)+chr((int(j)%26)+80)+('+' if (int(j)//26)+80 > (int(j)%26)+80 else '-') for j in ascii)  # 3. membuat 3 value dari setiap kode ascii dan menggabungkan semuanya
    print("Password terenkripsi : ", enkri)

def dekripsi(password):                                 # membuat sintax fungsi 'dekripsi'
  split_ps = [(password[k:k+3]) for k in range(0, len(password), 3)]  # split 3 value 
  # membuat 3 value kembali ke kode ascii lalu ke karakter asli dan menggabungkan semuanya
  dekri = ''.join(chr(((ord(list(z)[0])-80)*26)+(ord(list(z)[1])-80)) for z in split_ps)
  print("Password : ", dekri)
````

## **3b**. Apa output yang dihasilkan dari program tersebut jika input password adalah 'anakanakcerdas2020' ?

````python
password = input("Masukkan Password: ")
enkripsi(password)
     
````
